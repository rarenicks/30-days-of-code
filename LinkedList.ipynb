```python
# Linked List Common Questions

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# 1. Reverse Linked List (LeetCode 206)
# Intuition:
# Reverse the list by changing pointers.
# Maintain three pointers: prev, curr, and next.
# At each step, point curr.next to prev and move forward.
# This is a fundamental linked list operation often asked in interviews.
# Helps develop intuition around pointer manipulation and list traversal.
# Time: O(n), Space: O(1)
def reverseList(head):
    prev = None
    curr = head
    while curr:
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
    return prev

# 2. Reverse Between (LeetCode 92)
# Intuition:
# Reverse a sub-section of the linked list between two indices.
# Use a dummy node to navigate to the start, then reverse the sublist in-place.
# Head insertion technique makes the reversal efficient.
# Tests understanding of pointer juggling and sublist boundaries.
# Common for in-place mutation problems.
# Time: O(n), Space: O(1)
def reverseBetween(head, m, n):
    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    for _ in range(m - 1):
        prev = prev.next
    curr = prev.next
    for _ in range(n - m):
        tmp = curr.next
        curr.next = tmp.next
        tmp.next = prev.next
        prev.next = tmp
    return dummy.next

# 3. Detect Cycle (LeetCode 141)
# Intuition:
# Use Floyd’s Tortoise and Hare approach to detect if a cycle exists.
# Fast and slow pointers will meet if a cycle is present.
# Very efficient since it uses constant space.
# Great for understanding cyclic data structures.
# Time: O(n), Space: O(1)
def hasCycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# 4. Start of Cycle (LeetCode 142)
# Intuition:
# First detect if a cycle exists using two pointers.
# Then reset one pointer to the head and move both one step at a time.
# Where they meet is the entry point of the cycle.
# Builds on cycle detection and tests understanding of distance math.
# Time: O(n), Space: O(1)
def detectCycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    else:
        return None
    slow = head
    while slow != fast:
        slow = slow.next
        fast = fast.next
    return slow

# 5. Merge Two Sorted Lists (LeetCode 21)
# Intuition:
# Recursively or iteratively compare nodes from two sorted lists.
# Build the merged result by linking smaller nodes first.
# Tests handling of edge cases like empty lists.
# Core skill for merge sort and divide & conquer problems.
# Time: O(n + m), Space: O(1) iteratively
def mergeTwoLists(l1, l2):
    dummy = ListNode()
    tail = dummy
    while l1 and l2:
        if l1.val < l2.val:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next
    tail.next = l1 or l2
    return dummy.next

# 6. Odd Even Linked List (LeetCode 328)
# Intuition:
# Separate odd and even indexed nodes and then link them together.
# Maintain two lists and link the last odd node to the head of even list.
# Great test of pointer manipulation and understanding of node positions.
# Time: O(n), Space: O(1)
def oddEvenList(head):
    if not head or not head.next:
        return head
    odd = head
    even = head.next
    even_head = even
    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next
    odd.next = even_head
    return head

# 7. Palindrome Linked List (LeetCode 234)
# Intuition:
# Use two pointers to find the middle of the list.
# Reverse second half and compare both halves node by node.
# Requires strong understanding of reversing and list traversal.
# Time: O(n), Space: O(1)
def isPalindrome(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    prev = None
    while slow:
        tmp = slow.next
        slow.next = prev
        prev = slow
        slow = tmp
    while prev:
        if head.val != prev.val:
            return False
        head = head.next
        prev = prev.next
    return True

# 8. Remove N-th Node From End (LeetCode 19)
# Intuition:
# Use two pointers. Move the first pointer n steps, then move both until the first hits end.
# The second pointer will be at the (n+1)-th node from end. Remove next.
# A dummy node helps handle edge cases cleanly.
# Time: O(n), Space: O(1)
def removeNthFromEnd(head, n):
    dummy = ListNode(0, head)
    fast = slow = dummy
    for _ in range(n):
        fast = fast.next
    while fast.next:
        fast = fast.next
        slow = slow.next
    slow.next = slow.next.next
    return dummy.next

# 9. Intersection of Two Linked Lists (LeetCode 160)
# Intuition:
# Traverse both lists. When one ends, jump to the other’s head.
# This syncs up the lengths and ensures meeting at intersection point.
# Brilliant pointer redirection trick.
# Time: O(n + m), Space: O(1)
def getIntersectionNode(headA, headB):
    if not headA or not headB:
        return None
    a, b = headA, headB
    while a != b:
        a = a.next if a else headB
        b = b.next if b else headA
    return a

# 10. Copy List with Random Pointer (LeetCode 138)
# Intuition:
# Clone nodes and insert them next to originals. Fix random pointers.
# Then split the combined list back into original and clone.
# Very clever 3-pass trick to maintain O(1) space.
# Time: O(n), Space: O(1)
def copyRandomList(head):
    if not head:
        return None
    curr = head
    while curr:
        tmp = ListNode(curr.val)
        tmp.next = curr.next
        curr.next = tmp
        curr = tmp.next
    curr = head
    while curr:
        if curr.random:
            curr.next.random = curr.random.next
        curr = curr.next.next
    dummy = ListNode(0)
    copy, curr = dummy, head
    while curr:
        copy.next = curr.next
        curr.next = curr.next.next
        curr = curr.next
        copy = copy.next
    return dummy.next

# 11. Add Two Numbers (LeetCode 2)
# Intuition:
# Simulate digit-by-digit addition like manual arithmetic.
# Carry is handled across iterations.
# Simple but good test of linked list iteration and carry logic.
# Time: O(n), Space: O(1)
def addTwoNumbers(l1, l2):
    dummy = ListNode()
    curr = dummy
    carry = 0
    while l1 or l2 or carry:
        val = (l1.val if l1 else 0) + (l2.val if l2 else 0) + carry
        carry, val = divmod(val, 10)
        curr.next = ListNode(val)
        curr = curr.next
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    return dummy.next

# 12. Rotate List (LeetCode 61)
# Intuition:
# Connect list into a cycle, then break it at the right spot.
# Calculate effective rotations using k % length.
# Great for testing understanding of list length and pointer moves.
# Time: O(n), Space: O(1)
def rotateRight(head, k):
    if not head or not head.next or k == 0:
        return head
    old_tail = head
    length = 1
    while old_tail.next:
        old_tail = old_tail.next
        length += 1
    old_tail.next = head
    k %= length
    new_tail = head
    for _ in range(length - k - 1):
        new_tail = new_tail.next
    new_head = new_tail.next
    new_tail.next = None
    return new_head

# 13. Swap Nodes in Pairs (LeetCode 24)
# Intuition:
# Use iteration or recursion to swap every two adjacent nodes.
# Simple structure, but pointer updates are slightly tricky.
# Good practice for local two-node manipulation.
# Time: O(n), Space: O(1)
def swapPairs(head):
    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    while head and head.next:
        first = head
        second = head.next
        prev.next = second
        first.next = second.next
        second.next = first
        prev = first
        head = first.next
    return dummy.next

# 14. Reorder List (LeetCode 143)
# Intuition:
# Find middle, reverse second half, then merge both halves.
# Classic multi-step pointer problem testing full list manipulation.
# Time: O(n), Space: O(1)
def reorderList(head):
    if not head:
        return
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    prev = None
    curr = slow.next
    slow.next = None
    while curr:
        tmp = curr.next
        curr.next = prev
        prev = curr
        curr = tmp
    first, second = head, prev
    while second:
        tmp1, tmp2 = first.next, second.next
        first.next = second
        second.next = tmp1
        first, second = tmp1, tmp2

# 15. Split Linked List in Parts (LeetCode 725)
# Intuition:
# Determine size of each part based on total length and k.
# Distribute remainder 1-by-1 to the first few parts.
# Great for manipulating pointer-based partitions.
# Time: O(n), Space: O(k)
def splitListToParts(head, k):
    length = 0
    curr = head
    while curr:
        length += 1
        curr = curr.next
    width, rem = divmod(length, k)
    res = []
    curr = head
    for i in range(k):
        head = curr
        for j in range(width + (i < rem) - 1):
            if curr:
                curr = curr.next
        if curr:
            curr.next, curr = None, curr.next
        res.append(head)
    return res

# 16. Convert Binary Number in Linked List to Integer (LeetCode 1290)
# Intuition:
# Traverse list and build number left to right using bit shifting.
# Each step is result = result * 2 + node.val.
# Very efficient; tests numerical manipulation using lists.
# Time: O(n), Space: O(1)
def getDecimalValue(head):
    num = 0
    while head:
        num = num * 2 + head.val
        head = head.next
    return num

# 17. Delete Node in a Linked List (LeetCode 237)
# Intuition:
# Since we don’t have head, we copy the next node’s value to the current.
# Then skip over the next node.
# Very specific constraint makes this unique.
# Time: O(1), Space: O(1)
def deleteNode(node):
    node.val = node.next.val
    node.next = node.next.next
```
