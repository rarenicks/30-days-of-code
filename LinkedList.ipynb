```python
# Linked List Amazon SDE-2 Prep: Quick Code + Intuition

# 1. Reverse Linked List
# Intuition: Use 3 pointers: prev, curr, next
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseList(head):
    """
    Reverses a singly linked list by iteratively moving each node to the front.
    Maintain three pointers: prev, curr, and next to reverse the links.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    prev = None
    curr = head
    while curr:
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
    return prev

# 2. Reverse Between (m to n)
def reverseBetween(head, m, n):
    """
    Reverse the sublist from position m to n.
    Use pointer manipulation to reverse the section in-place.
    Insert elements at the front of the sublist using a dummy node.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    for _ in range(m - 1):
        prev = prev.next
    curr = prev.next
    for _ in range(n - m):
        tmp = curr.next
        curr.next = tmp.next
        tmp.next = prev.next
        prev.next = tmp
    return dummy.next

# 3. Detect Cycle
def hasCycle(head):
    """
    Detects if a linked list has a cycle using Floyd's Tortoise and Hare.
    Two pointers (slow and fast) move at different speeds.
    If they meet, a cycle exists.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# 4. Start of Cycle
def detectCycle(head):
    """
    Uses Floyd's algorithm to find the start of a cycle.
    Once cycle is detected, use a second pointer from head.
    The point where it meets slow again is the cycle start.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    else:
        return None
    slow = head
    while slow != fast:
        slow = slow.next
        fast = fast.next
    return slow

# 5. Merge Two Sorted Lists
def mergeTwoLists(l1, l2):
    """
    Merge two sorted linked lists by choosing the smaller node each time.
    Create a dummy node and iterate over both lists.
    Time Complexity: O(n + m)
    Space Complexity: O(1)
    """
    dummy = ListNode(-1)
    curr = dummy
    while l1 and l2:
        if l1.val < l2.val:
            curr.next = l1
            l1 = l1.next
        else:
            curr.next = l2
            l2 = l2.next
        curr = curr.next
    curr.next = l1 or l2
    return dummy.next

# 6. Middle of Linked List
def middleNode(head):
    """
    Find the middle of the list using slow and fast pointers.
    When fast reaches the end, slow is at the middle.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow

# 7. Remove N-th Node From End
def removeNthFromEnd(head, n):
    """
    Use two pointers to remove the nth node from the end.
    Advance fast by n nodes, then move both pointers.
    When fast reaches end, slow is before the target.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    dummy = ListNode(0, head)
    slow = fast = dummy
    for _ in range(n):
        fast = fast.next
    while fast.next:
        slow = slow.next
        fast = fast.next
    slow.next = slow.next.next
    return dummy.next

# 8. Reorder List
def reorderList(head):
    """
    Reorder the list by reversing the second half and merging.
    Find middle, reverse second half, merge both halves.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    if not head:
        return
    slow, fast = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    prev, curr = None, slow.next
    slow.next = None
    while curr:
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
    first, second = head, prev
    while second:
        tmp1, tmp2 = first.next, second.next
        first.next = second
        second.next = tmp1
        first, second = tmp1, tmp2

# 9. Palindrome Linked List
def isPalindrome(head):
    """
    Use fast and slow to find mid, reverse second half.
    Compare first and second halves node by node.
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    slow, fast = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    prev, curr = None, slow
    while curr:
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
    left, right = head, prev
    while right:
        if left.val != right.val:
            return False
        left = left.next
        right = right.next
    return True

# 10. Intersection of Two Lists
def getIntersectionNode(headA, headB):
    """
    Traverse both lists; switch heads when reaching the end.
    This aligns the pointers to meet at intersection if exists.
    Time Complexity: O(n + m)
    Space Complexity: O(1)
    """
    a, b = headA, headB
    while a != b:
        a = a.next if a else headB
        b = b.next if b else headA
    return a

# 11. Copy List with Random Pointer
class Node:
    def __init__(self, val=0, next=None, random=None):
        self.val = val
        self.next = next
        self.random = random

def copyRandomList(head):
    """
    Clone the list using interleaved copy and original nodes.
    Use a second pass to assign random pointers.
    Time Complexity: O(n)
    Space Complexity: O(1) extra
    """
    if not head:
        return None
    curr = head
    while curr:
        copy = Node(curr.val)
        copy.next = curr.next
        curr.next = copy
        curr = copy.next
    curr = head
    while curr:
        if curr.random:
            curr.next.random = curr.random.next
        curr = curr.next.next
    curr = head
    new_head = head.next
    while curr:
        copy = curr.next
        curr.next = copy.next
        copy.next = copy.next.next if copy.next else None
        curr = curr.next
    return new_head

# 12. Add Two Numbers
def addTwoNumbers(l1, l2):
    """
    Simulate addition using carry and traverse both lists.
    Create a new node for each digit.
    Time Complexity: O(n)
    Space Complexity: O(n)
    """
    dummy = ListNode(0)
    curr = dummy
    carry = 0
    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        total = val1 + val2 + carry
        carry = total // 10
        curr.next = ListNode(total % 10)
        curr = curr.next
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    return dummy.next

# 13. Add Two Numbers II (Forward Order)
def addTwoNumbersII(l1, l2):
    """
    Use stacks to reverse the order.
    Pop and add digits, build result backward.
    Time Complexity: O(n)
    Space Complexity: O(n)
    """
    s1, s2 = [], []
    while l1:
        s1.append(l1.val)
        l1 = l1.next
    while l2:
        s2.append(l2.val)
        l2 = l2.next
    carry = 0
    head = None
    while s1 or s2 or carry:
        val1 = s1.pop() if s1 else 0
        val2 = s2.pop() if s2 else 0
        total = val1 + val2 + carry
        carry = total // 10
        node = ListNode(total % 10)
        node.next = head
        head = node
    return head

# 14. Flatten Multilevel Doubly Linked List
def flatten(head):
    """
    Use a stack to flatten child lists in-place.
    Attach children before moving forward.
    Time Complexity: O(n)
    Space Complexity: O(n)
    """
    if not head:
        return head
    stack, curr = [], head
    while curr:
        if curr.child:
            if curr.next:
                stack.append(curr.next)
            curr.next = curr.child
            curr.child.prev = curr
            curr.child = None
        elif not curr.next and stack:
            curr.next = stack.pop()
            curr.next.prev = curr
        curr = curr.next
    return head

# 15. LRU Cache
from collections import OrderedDict
class LRUCache:
    """
    Use OrderedDict to maintain order of access.
    Move key to end on access, remove least recent if full.
    Time Complexity: O(1) per operation
    """
    def __init__(self, capacity: int):
        self.cache = OrderedDict()
        self.capacity = capacity
    def get(self, key: int) -> int:
        if key in self.cache:
            self.cache.move_to_end(key)
            return self.cache[key]
        return -1
    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            self.cache.move_to_end(key)
        self.cache[key] = value
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)

# 16. Sort Linked List (Merge Sort)
def sortList(head):
    """
    Sort the list using merge sort recursively.
    Split, sort and merge the two halves.
    Time Complexity: O(n log n)
    Space Complexity: O(log n) for recursion
    """
    if not head or not head.next:
        return head
    slow, fast = head, head.next
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    mid = slow.next
    slow.next = None
    left = sortList(head)
    right = sortList(mid)
    return mergeTwoLists(left, right)

# 17. Split List into K Parts
def splitListToParts(head, k):
    """
    Count total nodes and divide equally among k parts.
    Distribute extra nodes to first few parts.
    Time Complexity: O(n)
    Space Complexity: O(k)
    """
    length = 0
    curr = head
    while curr:
        length += 1
        curr = curr.next
    part_len, extra = divmod(length, k)
    result = []
    curr = head
    for _ in range(k):
        part_head = curr
        for _ in range(part_len - 1 + (extra > 0)):
            if curr:
                curr = curr.next
        if curr:
            next_part = curr.next
            curr.next = None
            curr = next_part
        result.append(part_head)
        extra -= 1
    return result
```
